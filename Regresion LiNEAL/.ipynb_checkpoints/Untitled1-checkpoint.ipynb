## MODELO DE REGRESION LINEAL  x=radio y=ventas

```python
X=datos[["Radio"]]
```

```python
y=datos["Sales"]
```

#### APLICACION DE MINIMOS CUADRADOS

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X,y)
```

#### PARAMETROS OBTENIDOS PARA EL MODELO

```python
print(f'B0={modelo.intercept_}')
```

```python
print(f'B0={modelo.coef_[0]}')
```

```python
print(f'y`={modelo.coef_[0]} + {modelo.intercept_} x')
```

#### PREDICCIONES

```python
predicciones = modelo.predict(X)
```

```python
plt.scatter(y,predicciones)
```

```python
sns.regplot(x=y,y=predicciones,data=datos)
```

#### METRICAS DE EVALUACION

```python
from sklearn import metrics
```

```python
RSS = print(f'RSS: {metrics.mean_squared_error(y,predicciones)*200}') #SUMA DE LOS ERRORES AL CUADRADO
```
