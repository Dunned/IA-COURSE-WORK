```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

## FUNCION PARA MOSTRAR IMAGENES

```python
def mostrar_imagen(titulo,imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(titulo)
    plt.grid(False)
    plt.show
```

## CREAMOS LIENZOS CON NUMPY

```python
np.zeros((2,2),dtype='uint8')
```

```python
# inicializar el lienzo con una imagen de 300x300 pixeles con 3
# canales (Rojo, Verde y Az√∫l) con un fondo negro
canvas = np.zeros((300, 300, 3), dtype="uint8") # uint 8 bits = sin signo de 0-255
```

```python
mostrar_imagen('Lienzo',canvas)
```

```python
# dibujar una l√≠nea verde desde la esquina superior izquierda del
# lienzo hasta la esquina inferior derecha
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar una l√≠nea roja con 3 pixeles de grosor desde la esquina
# superior derecha a la esquina inferior izquierda
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar una l√≠nea azul con 10 pixeles de grosor desde la mitad
# de arriba hasta abajo
blanco = (255, 255, 255)
cv2.line(canvas, (150, 0), (150, 300), blanco, 3)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar un cuadrado verde de 50x50 pixeles, comenzando en 10,10 y
# terminando en 60,60
cv2.rectangle(canvas, (10, 10), (60, 60), green)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar otro rect√°ngulo, uno rojo con 5 pixeles de grosor
cv2.rectangle(canvas, (10, 200), (200, 225), red, 5)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar un rect√°ngulo y rellenarlo (az√∫l y rellenado )
verde = (0, 255, 0) #B,G,R
cv2.rectangle(canvas, (200, 50), (225, 125), verde, -1) # EN CORDENADAS X , Y / 
mostrar_imagen("Canvas", canvas)
```

```python
# reinicializar el canvas como un arreglo vac√≠o, calcular el
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
# centro (x,y) del canvas
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
```

```python
red = (0, 0, 255)
```

```python
# bucle para incrementar el radio, desde 25 pixeles hasta 150 pixeles
# en incrementos de 25
for r in range(0, 175, 25):
    # draw a white circle with the current radius size
    cv2.circle(canvas, (centerX, centerY), r, red)
    # mostrar el dibujo
mostrar_imagen("Canvas", canvas)
```

```python
# reinicializar el canvas nuevamente
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
# dibujemos 25 circulos
for i in range(0, 25):
# generar un radio aleatorio entre 5 y 200, generar un color
# aleatorio, y luego escoger aleatoriamente un punto en el
# lienzo donde sera dibujado el c√≠rculo
    radius = np.random.randint(5, high=200)
    color = np.random.randint(0, high=256, size=(3,)).tolist()
    pt = np.random.randint(0, high=300, size=(2,))
    # dibujar el circulo aleatorio
    cv2.circle(canvas, tuple(pt), radius, color, -1)
    # mostrar el dibujo
mostrar_imagen("Canvas", canvas)
```

# DIBUJAMOS

```python
args = {
"image": "paisaje.jpeg"
}
```

```python
# cargar la imagen
image = cv2.imread(args["image"])
# dibujar un circulo, dos circulos rellenos y un rect√°ngulo
# Dibuja una l√≠nea divisoria vertical negra en el centro
cv2.line(image, (300, 0), (300, 500), (0, 0, 0), 2)
cv2.line(image, (300, 250), (700, 250), (0, 0, 0), 2)
cv2.circle(image, (200, 300), 75, (0, 0, 255), 5)
cv2.rectangle(image, (20, 240), (100, 360), (255, 0, 0), thickness=3)

triangles = np.array([
    [(200, 240), (145, 333), (255, 333)],
    [(60, 180), (20, 237), (100, 237)]])
cv2.fillPoly(image, triangles, (0, 255, 0))


np.random.seed(42)
n_pts = 30
pts_x = np.random.randint(310, 590, n_pts)
pts_y = np.random.randint(160, 390, n_pts)
pts = zip(pts_x, pts_y)


for pt in pts:
    pt_color = [int(c) for c in np.random.randint(0, 255, 3)]
    cv2.circle(image, pt, 3, pt_color, 5)
    
cv2.putText(image,
            'Imagen Experimental',
            (5, 15),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0, 0, 0),
            1)

# mostrar la imagen
mostrar_imagen("Output", image)
```
