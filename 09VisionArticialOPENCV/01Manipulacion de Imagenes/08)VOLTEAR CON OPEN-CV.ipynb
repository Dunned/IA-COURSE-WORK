## IMPORTAMOS LIBRERIAS

```python
import matplotlib.pyplot as plt
import imutils
import cv2
```

## FUNCION PARA MOSTRAR IMAGEN

```python
def mostrar_imagen(title, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## IMPLEMENTACIÓN DE VOLTEAR IMAGENES CON OPEN CV

```python
# CARGAMOS EL NOMBRE DE LA IMAGEN EN EL DICCIONARIO ARGS
```

```python
args={
    'imagen':'mars.jpeg'
}
```

```python
# CARGAMOS LA IMAGEN Y LA MOSTRAMOS
```

```python
imagen=cv2.imread(args['imagen'])
mostrar_imagen('Imagen Original de Mars',imagen)
```

```python
#VOLTEAREMOS LA IMAGEN HORIZONTALMENTE
```

```python
volteado=cv2.flip(imagen,1) # 1=Horizontalmente , 0=Verticalmente
mostrar_imagen('Volteado Horizontalmente',volteado)
```

```python
#VOLTEAREMOS LA IMAGEN VERTICALMENTE
```

```python
volteado=cv2.flip(imagen,0) # 1=Horizontalmente , 0=Verticalmente
mostrar_imagen('Volteado Verticalmente',volteado)
```

```python
#VOLTEAMOS IMAGEN EN AMBOS EJES
```

```python
volteado=cv2.flip(imagen,-1) #-1=En ambos Ejes
mostrar_imagen('Volteado ambos ejes',volteado)
```
