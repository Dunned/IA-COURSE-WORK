##### IMPORTAMOS LIBRERIAS

```python
from matplotlib import pyplot as plt
import cv2
```

###### FUNCION PARA MOSTRAR IMAGENES

```python
def mostrarImagen(nombreImagen=None,titulo=None):
 imagen1=cv2.imread(nombreImagen)
 imagen1=cv2.cvtColor(imagen1,cv2.COLOR_BGR2RGB)
 plt.imshow(imagen1)
 plt.title(titulo)
 plt.grid(False)
 plt.show
```

```python
mostrarImagen("beach.png",'Imagen del Oceano')
```

```python
mostrarImagen("coins.png",'Monedas')
```

```python
mostrarImagen("celular.jpeg",'Celular Maqueta')
```

```python
mostrarImagen('daronco.jpeg','El mejor arbitro del mundo')
```
