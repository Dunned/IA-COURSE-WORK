## IMPORTAMOS LIBRERIAS

```python
import matplotlib.pyplot as plt
import imutils
import cv2
```

## FUNCION PARA MOSTRAR IMAGEN

```python
def mostrar_imagen(title, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## Implementación de la rotación con OpenCV

```python
args = {
"imagen" :"unmsm.png"
}
```

```python
#Cargar la imagen y mostrarla
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

```python
# Determinar el centro de la imagen y
# calcular el centro de la imagen
(h,w) = imagen.shape[0:2]
(cX, cY) = (w//2,h//2)
print((cX,cY))
```

```python
# rotar la imagen 45° alrededor del centro de la imagen
M = cv2.getRotationMatrix2D((cX,cY),45,1.5) # (cx,cy CENTRO DE LA IMAGEN),GRADOS,ESCALA
rotado = cv2.warpAffine(imagen,M,(w,h))
mostrar_imagen("Rotado 45°",rotado)
```

```python
# rotar la imagen 45° alrededor de LA CORDENADA X=10 Y=10
M = cv2.getRotationMatrix2D((10,10),45,1) # (cx,cy CENTRO DE LA IMAGEN),GRADOS,ESCALA
rotado = cv2.warpAffine(imagen,M,(w,h))
mostrar_imagen("Rotado 45°",rotado)
```

```python
# rotar la imagen -90° alrededor del centro de la imagen
M = cv2.getRotationMatrix2D((cX,cY),-90,0.2)
rotado = cv2.warpAffine(imagen,M,(w,h))
mostrar_imagen("Rotado -90°",rotado)
```

## UTILIZANDO IMUTILS

```python
rotado=imutils.rotate(imagen,-90)
mostrar_imagen('Rotado 90',rotado)
```

## ROTAR SIN CORTAR IMAGEN (REDIMENSIONAR)

```python
rotado=imutils.rotate_bound(imagen,-100)
mostrar_imagen('Rotado 100',rotado)
```
