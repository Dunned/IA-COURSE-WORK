```python
from matplotlib import pyplot as plt
import cv2
```

## Cargamos la Imagen

```python
imagen=cv2.imread("Poker02.jpeg") #Devuelve una Matriz  // RGB ORIGINAL CAMBIA A -> BGR (IMAGEN DE OPENCV)
```

```python
imagen.shape # detalla valores -> (altura,ancho,ultimo valor = 3 capas/canales [B,G,R] )
```

```python
plt.imshow(imagen) # mostramos la imagen como RGB , pero en realidad es BGR , por eso esta opaco
```

## CONVERTIMOS IMAGEN DE FORMATO BGR A RGB

```python
imagen=cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB) #Convertimos imagen a RGB
```

```python
plt.imshow(imagen) # mostramos la imagen / AHORA SI SALE CLARO!
```

## FUNCION PARA MOSTRAR IMAGENES

```python
def mostrarImagen(nombreImagen=None,titulo=None):
    imagen1=cv2.imread(nombreImagen)
    imagen1=cv2.cvtColor(imagen1,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen1)
    plt.title(titulo)
    plt.grid(False)
    plt.show
```

```python
mostrarImagen("opencv_logo.png",'ImagenEjemplo')
```

## GUARDARMOS IMAGEN

```python
cv2.imwrite("NUEVONOMBRE22.jpg",cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)) 
#guardamos la imagen en la direccion de la carpeta , en el sistema color
```

## FUNCION PARA GUARDAR IMAGEN

```python
def guardarImagen(nombre,imagenEntrada): #IMAGEN LEIDA CON imread
    cv2.imwrite(nombre,cv2.cvtColor(imagenEntrada,cv2.COLOR_BGR2RGB)) 
```
