## IMPORTAMOS LIBRERIAS

```python
import matplotlib.pyplot as plt
import imutils
import cv2
```

## FUNCION PARA MOSTRAR IMAGEN

```python
def mostrar_imagen(title, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## Implementaci√≥n de la Redimension con OpenCV

```python
args = {
"imagen" :"daronco.jpeg"
}
```

```python
#Cargar la imagen y mostrarla
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

```python
print(imagen.shape)
```

```python
# Redimensionar la imagen para tener 200 px de ancho
# haciendo que la otra dimensi√≥n tenga un tama√±o
# apropiado y no se distorsione
# proporci√≥n de reducci√≥n de la imagen
r = 200.0/imagen.shape[1] # ?CUANTO ES % ES 200 DEL ANCHO?
print(f'{r}%')
```

```python
# nueva dimensi√≥n de la imagen
dim = (200, int(imagen.shape[0]*r)) # alto por r para no alterar la imagen
print('Nueva dimensiones')
print(f'Ancho:{dim[0]}')
print(f'Alto:{dim[1]}')
```

```python
# redimensionar la imagen
redimensionado = cv2.resize(imagen,dim,cv2.INTER_AREA) # reducir=INTER_AREA - AGRANDAR = INTER_CUBIC
mostrar_imagen("Redimensionado a un ancho de 200 pixeles",redimensionado)
```

## UTILIZANDO IMUTILS

```python
# redimensionamiento de la imagen a un alto de 300 pixeles

# manteniemdo la proporci√≥n de la imagen
r = 300.0/imagen.shape[1]
# nueva dimensi√≥n de la imagen
dim = (300,int(imagen.shape[1]*r))
# redimensionar la imagen
redimensionado = cv2.resize(imagen,dim,cv2.INTER_CUBIC)
mostrar_imagen("Redimensionado a un alto de 300 pixeles",redimensionado)
```

## GUARDAMOS IMAGEN

```python
cv2.imwrite('Redimensiona.jpg',redimensionado) ## COMO ESTA EN BGR LO TRANSFORMA A RGB
```

```python
# VAMOS A REDIMENSIONAR LA IMAGEN A UN ALTO DE 500 PIXELES , MANTENIENDO LA PROPORCION
```

```python
r=500.0/imagen.shape[0]
#nueva dimension de la imagen
dim=(int(imagen.shape[1]*r),500)
#redimensionar la imagen
redimensionado=cv2.resize(imagen,dim,cv2.INTER_AREA)
mostrar_imagen('Redimensionando a un alto de 500 pixeles',redimensionado)
```

```python
redimensionado=imutils.resize(imagen,width=90) #width=x,height=x
mostrar_imagen('Redumensionado a un ancho de 90',redimensionado)
```

## Comparaci√≥n de m√©todos de interpolaci√≥n con OpenCV

```python
metodos = [("cv2.INTER_NEAREST",cv2.INTER_NEAREST),
("cv2.INTER_LINEAR",cv2.INTER_LINEAR),
("cv2.INTER_AREA",cv2.INTER_AREA),
("cv2.INTER_CUBIC",cv2.INTER_CUBIC),
("cv2.INTER_LANCZOS4",cv2.INTER_LANCZOS4)]
```

```python
for (nombre,metodo) in metodos:
    print("M√©todo:{}".format(nombre))
    redimensionado = imutils.resize(imagen,width=1980,inter=metodo)
    mostrar_imagen("Metodo: {}".format(nombre),redimensionado)
```
