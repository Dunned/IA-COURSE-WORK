```python
from matplotlib import pyplot as plt
import argparse
import cv2
```

# FUNCION

```python
def mostrar_imagen(titulo,imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(titulo)
    plt.grid(False)
    plt.show
```

## OBTENCION Y MODIFICACION DE PIXELES

```python
args = {
"imagen": "peru.jpeg"
}
```

```python
# cargar la imágen, graba sus dimensiones (ancho y alto),
# y luego muestra la imagen original
imagen = cv2.imread(args["imagen"])
(h, w) = imagen.shape[:2]
mostrar_imagen("Original", imagen)
print(f'Ancho IMAGEN: {w}')
print(f'Alto IMAGEN: {h}')
```

```python
# Las imagenes son arreglos Numpy -- con el origen (0,0) ubicado en
# la esquina superior izquierda de la imagen
(b, g, r) = imagen[0, 0]
print("Pixel en (0, 0) - Rojo: {}, Verde: {}, Azúl: {}".format(r, g, b))
```

```python
# accede al pixel ubicado en x=400, y=300
(b, g, r) = imagen[300, 400]
print("Pixel en (x=400, y=300) - Rojo: {}, Verde: {}, Azúl: {}".format(r, g, b))
```

```python
# actualizar el pixel en (400, 300) and establecerlo a rojo
imagen[300, 400] = (0, 0, 255)
(b, g, r) = imagen[300, 400]
print("Pixel en (x=800, y=400) - Rojo: {}, Verde: {}, Azúl: {}".format(r, g, b))
mostrar_imagen('Pixel cambiado',imagen)
```

```python
# calcula el centro de la imagen, simplemente es el ancho y alto
# dividido por dos
(cX, cY) = (w // 2, h // 2)
print(f'Centro: x={cX},y={cY}')
```

# ESCOGER PARTE DE IMAGEN

```python
tl = imagen[0:cY, 0:cX]
```

```python
mostrar_imagen("Esquina superior izquierda", tl)
```

```python
tr = imagen[0:cY, cX:w]
br = imagen[cY:h, cX:w]
bl = imagen[cY:h, 0:cX]
mostrar_imagen("Esquina superior derecha", tr)
```

```python
mostrar_imagen("Esquina inferior derecha", br)
```

```python
mostrar_imagen("Esquina inferior izquierda", bl)
```

```python
#esquina superrior izquierda color verde
imagen[0:cY,0:cX]=(0,255,0)
mostrar_imagen('Actualizadfa Verde',imagen)
```
