## IMPORTAMOS LIBRERIAS

```python
import matplotlib.pyplot as plt
import imutils
import numpy as np
import cv2
```

## FUNCION PARA MOSTRAR IMAGEN

```python
def mostrar_imagen(title=None, imagen=None):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## IMPLEMENTACIÓN DEL RECORTE DE IMAGENES CON OPENCV Y NUMPY

```python
#GUARDAMOS LA IMAGEN EN UN DICCIONARIO CON CLAVE 'IMAGEN'
args={
    'imagen':'peru2.jpeg'
}
```

```python
#CARGAMOS LA IMAGEN Y LA MOSTRAMOS
imagen=cv2.imread(args['imagen'])
mostrar_imagen('Imagen Original Seleccion Peruana',imagen)
```

```python
# -> 0:200 DEFINE EL ALTO DE LA IMAGEN
# ->0:800 DEFINE EL ANCHO DE LA IMAGEN
recortado=imagen[0:200,0:800]    #[y=0->y=200,x=0->x=800]
mostrar_imagen('Imagen Recortada',recortado)
```

```python
#INTENTEMOS RECORTAR LA CARA DEL OREJA FLORES
```

```python
#EMPEZAMOS A RECORTAR CON NUMPY
recortado=imagen[70:230,210:430]    #[y=70->y=230,x=210->x=430]
mostrar_imagen('Cara de Edison Flores Recortada',recortado)
```

```python
#INTENTEMOS RECORTAR LA CARA DEL PEDRO AQUINO
```

```python
#EMPEZAMOS A RECORTAR CON NUMPY
recortado=imagen[10:200,500:700]    #[y=10->y=200,x=500->x=700]
mostrar_imagen('Cara de Pedro Aquino Recortada',recortado)
```
